```python
def encode_text(text):
    # Tokenize and encode the text input
    input_ids = torch.tensor([tokenizer.encode(text, add_special_tokens=True)])
    return input_ids

def encode_data_text(data):
    # Tokenize and encode the 'text' column in the DataFrame
    input_ids = torch.tensor([tokenizer.encode(text, add_special_tokens=True) for text in data['text']])
    return input_ids

# Example usage:
text_input = "This is a sample text for comparison."
encoded_input = encode_text(text_input)

# Assuming your DataFrame is named 'data'
data_encoded = encode_data_text(data)

def get_bert_embeddings(input_ids):
    with torch.no_grad():
        model.eval()
        outputs = model(input_ids)
    return outputs.last_hidden_state

# Example usage:
text_input_embeddings = get_bert_embeddings(encoded_input)
data_embeddings = get_bert_embeddings(data_encoded)

from torch.nn.functional import cosine_similarity

def calculate_cosine_similarity(input_embeddings, data_embeddings):
    return cosine_similarity(input_embeddings, data_embeddings)

# Example usage:
similarities = calculate_cosine_similarity(text_input_embeddings, data_embeddings)

def get_most_similar_texts(similarities, data):
    sorted_indices = similarities.argsort(descending=True)
    return data.iloc[sorted_indices]

# Example usage:
most_similar_texts = get_most_similar_texts(similarities, data)

```

```python
def space_newline(input_string):
    words = input_string.replace('\n', ' ').split(' ')
    filtered_words = [word for word in words if word]
    return ' '.join(filtered_words)
```
